```python
import pandas as pd

data = pd.read_csv("AB_NYC_2019.csv")
data.head()
```

## **Understanding the Dataset**

Before cleaning the dataset, it is important to understand its structure. This includes checking the number of rows, columns, and the data types of each column.

This step helps ensure **data integrity**, meaning the dataset is accurate and consistent.


```python
data.info()
```

## Statistical Summary

Statistical summary helps us understand numerical columns in the dataset. It shows values such as mean, minimum, maximum, and standard deviation.


```python
data.describe()
```

## Missing Data Handling

Missing values occur when some data entries are not recorded. Handling missing data is important because it can affect the accuracy of analysis.

We will identify and handle missing values in the dataset.


```python
data.isnull().sum()
```

```python
data = data.fillna(method="ffill")
```

## Duplicate Removal

Duplicate records occur when the same data entry appears multiple times. These duplicates can distort analysis and should be removed to maintain data uniqueness.


```python
data.duplicated().sum()
```

## Data Standardization

Standardization ensures that data follows a consistent format. This improves readability and ensures accurate analysis.


```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.strip()
```

## Outlier Detection

Outliers are extreme values that differ significantly from other observations in the dataset. Detecting outliers is important because they can affect statistical analysis and model performance.


```python
sns.boxplot(x=data["price"])
plt.show()
```

```python
data.to_csv("cleaned_airbnb_data.csv", index=False)
```
