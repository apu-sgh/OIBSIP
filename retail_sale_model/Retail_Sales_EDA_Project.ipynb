# **Insights**

The dataset contains 260 menu items across multiple food categories
such as Breakfast, Beef & Pork, Chicken & Fish, Salads, Desserts, and Smoothies & Shakes.

Items in the Smoothies & Shakes category tend to have the highest calorie values compared to most other categories.

There is a strong positive relationship between calories and total fat, meaning menu items with higher fat generally contain more calories.

Foods with higher sugar levels also show higher calorie values, especially beverages and dessert items.

Lighter options such as salads and some breakfast items generally have lower calories and fat compared to burgers and desserts.

Nutritional components such as sodium and cholesterol are significantly higher in fried and meat-based menu items.

# **Recommendations**

McDonald's can introduce more low-calorie and low-fat menu options for health-conscious customers.

Reducing sugar content in beverages and desserts could help improve the nutritional quality of the menu.

Providing clear nutritional labels can help customers make healthier food choices.

Promoting healthier menu items such as salads and grilled foods can encourage balanced eating habits.

Limiting high sodium and high fat ingredients may help improve overall nutritional value.

```python
import pandas as pd

retail = pd.read_csv("retail_sales_dataset.csv")

retail.head()
```

```python
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
```

```python
retail["Total_Sales"] = retail["Quantity"] * retail["Price per Unit"]

retail.head()
```

```python
import matplotlib.pyplot as plt

category_sales = retail.groupby("Product Category")["Total_Sales"].sum()

category_sales.plot(kind="bar")
plt.title("Total Sales by Product Category")
plt.ylabel("Sales")
plt.show()
```

```python
import seaborn as sns

sns.countplot(x="Gender", data=retail)
plt.title("Customer Gender Distribution")
plt.show()
```

```python
sns.histplot(retail["Age"], bins=20)
plt.title("Customer Age Distribution")
plt.show()
```

```python
retail["Date"] = pd.to_datetime(retail["Date"])

sales_trend = retail.groupby("Date")["Total_Sales"].sum()

sales_trend.plot(figsize=(10,5))
plt.title("Sales Trend Over Time")
plt.ylabel("Total Sales")
plt.show()
```

```python
numeric_data = retail.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_data.corr(), annot=True)
plt.title("Retail Sales Correlation Heatmap")
plt.show()
```

# **Insights**

The retail dataset contains customer information such as age, gender, and product categories.

Some product categories contribute more to total sales compared to others.

The majority of customers fall within certain age groups, indicating a target demographic.

Sales vary across different dates, showing fluctuations in daily revenue.

Quantity purchased and price per unit significantly influence total sales.

# **Recommendations**

Focus marketing efforts on the most profitable product categories.

Target advertisements toward the dominant customer age group.

Offer promotions during low-sales periods to increase revenue.

Maintain proper inventory for high-demand products.
